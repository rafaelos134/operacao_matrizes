```python
# -*- coding: utf-8 -*-
import numpy as np
  
def subtracao_linhas(linha, ab):
    for x in range(ab.shape[1]):
        if ab[x,linha] != 0 and not np.array_equal(ab[linha],ab[x]) and x > linha:
                ab[x] = ab[x] - ab[linha]*ab[x,linha]    
    return ab
            
def divisao_pivo(linha,ab,det):
    if ab[linha,linha] != 1 and ab[linha,linha] != 0:
        det *= ab[linha,linha]
        ab[linha] = ab[linha]*(1/ab[linha,linha])
    return det,ab

def matriz_cofatora(matriz):
    n = matriz.shape[0]
    cofatores = np.zeros_like(matriz, dtype=float)
    for i in range(n):
        for j in range(n):
            submatriz = matriz[np.arange(n) != i][:, np.arange(n) != j]
            cofatores[i, j] = ((-1) ** (i + j)) * np.linalg.det(submatriz)
    adj = cofatores.T
    
    return adj




def main(A):
    
    ab = A
    matriz = []
    det = 1

    adj_AB = matriz_cofatora(ab)
    # pesquisa todas as linhas com o primiro pivor 
    for x in range(ab.shape[1]):
        if ab[x,0] == 1:
            matriz.append(x)

    # garante que o primeiro lemento é um pivor 
    if ab[0,0] != 1 and matriz:
        ab[0], ab[matriz[0]] = ab[matriz[0]].copy(), ab[0].copy()
        det*=-1

    shape = ab.shape[1]

    for x in range(0,int(shape)-1,1):
        ab = subtracao_linhas(x, ab)
        det,ab = divisao_pivo(x+1, ab,det)
        
    print("\nMatriz Escalonada de A|B:")
    print(np.round(ab, decimals=4))
    print("\nMatriz Adjunta de A|B:")
    print(adj_AB)
    print("\nMatriz transposta de A|B:")
    print(np.round(adj_AB*(1/det), decimals=4))
    print("\nDeterminante pelo metodo de Gauss-Jordan da matriz A|B:",round(det))
  
    

matriz = np.array([[1,10,4,-3,-1,5],[2,2,0,4,6,-14],[3,-1,-1,-1,10,7],[13,-11,5,4,-3,0],[5,-4,-4,3,-1,0],[0,0,6,9,1,-5]],dtype = float)
main(matriz)
```
